```python
i = 0
for i in range(len(os.listdir('letters_v1.2/'))):
    source = 'letters_v1.2/' + str(i)
    dest = 'letters_v1.2/' + str(i) +'.txt'
    os.rename(source,dest)
```

```python
letters = []
for i in range(78):
    filename = 'letters_v1.2/' + str(i)+'.txt'
    text_file = open(filename).read()
    letters.append(text_file)
print('loaded all')
```

```python
len(letters)
```

```python
len(letters[1])
```

```python
max_len = 0
for letter in letters:
    if len(letter) > max_len:
        max_len = len(letter)
print('Letter with max length has ' + str(max_len) +' characters')
```

```python
import matplotlib.pyplot as plt

x = []
y = []
for i in range(len(letters)):
    x.append(i)
    y.append(len(letters[i]))
    #print('letter '+str(i)+' length eduthu')

plt.plot(x,y)
plt.show()
```

```python
y.sort()
```

```python
y_greater_than_2000 = [i for i in y if i >= 2000]
```

```python
y_greater_than_2000
```

```python
pos = []
for i in range(len(y)):
    for j in range(len(y_greater_than_2000)):
        if y[i] == y_greater_than_2000[j]:
            pos.append(i)
print(pos)
```

```python
for positions in pos:
    print(y[positions])
```
